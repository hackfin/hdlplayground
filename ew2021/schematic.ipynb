```python
from graphviz import Digraph

toggle_code()

def schematic(which = 0):

    dot = Digraph("Schematic")
    dot.attr('node', shape='parallelogram')
    dot.node('A', 'MyHDL' ); dot.node('V', 'Vendor Sim Verilog'); 
    dot.attr('node', shape='box', style='filled', color='lightgrey')
    dot.node('M', 'Synthesizeable Verilog');

    dot.node('B', 'VHDL'); dot.node('C', 'Verilog'); dot.node('N', 'Synthesized net list')

    dot.attr('node', shape='ellipse', color='lightblue')
    dot.node("G", "GHDL plugin")
    dot.node('Y', 'Yosys' ); 
    dot.attr('node', shape='doubleoctagon')

    if which:
        edges = ['AB', 'AC', 'BG', 'VY', 'GY', 'YN', 'YM',  'CY']
        dot.edge('A', 'Y', color='red')
    else:
        edges = ['AB', 'AC', 'BG', 'VY', 'GY', 'YN', 'YM',  'CY']
    dot.edges(edges)
    
    return dot
```

```python
schematic(1)
```
